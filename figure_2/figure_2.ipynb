# Figure 2A

Compare total, spine, and "shaft" synapse density on Pyr and PV neurons

```python
# Aggregate Pyr data
paths = get_synfiles('../data/synfiles/pyr')

total_agg = []
spine_agg = []
shaft_agg = []

for path in paths:
    data = load_synfile(path)
    total, spine, shaft = calc_true_dens(data)

    total_agg.append(np.mean(total))
    spine_agg.append(np.mean(spine))
    shaft_agg.append(np.mean(shaft))

Pyr_total = total_agg

fig, ax = plt.subplots(1)
ax.set_title("Pyr")
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)

# Print out Pyr statistics
for condition, agg in zip(
    ["Total", "Spine", "Shaft"], [total_agg, spine_agg, shaft_agg]
):
    print("{}: Pyr {}+-{}".format(condition,np.mean(agg), scipy.stats.sem(agg)))
    ax.bar(condition, np.mean(agg), yerr=scipy.stats.sem(agg), capsize=5,color='blue')
plt.show()
```

```python
# Aggregate PV data
pv_path = '../data/synfiles/pv'
paths = get_synfiles(pv_path)

total_agg = []
spine_agg = []
shaft_agg = []

for path in paths:
    data = load_synfile(path)
    total, spine, shaft = calc_true_dens(data)
    total_agg.append(np.mean(total))
    spine_agg.append(np.mean(spine))
    shaft_agg.append(np.mean(shaft))

PV_total = total_agg

fig, ax = plt.subplots(1)
ax.set_title("PV")
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)

for condition, agg in zip(
    ["Total", "Spine", "Shaft"], [total_agg, spine_agg, shaft_agg]
):
    print("{}: PV {}+-{}".format(condition, np.mean(agg), scipy.stats.sem(agg)))
    ax.bar(condition, np.mean(agg), yerr=scipy.stats.sem(agg), capsize=5,color='red')

plt.show()
```

```python
print("T-Test: {}".format(scipy.stats.ttest_ind(Pyr_total, PV_total)))
```

# Figure 2B

Comparison of raw PSD-95 fluorescence between depth-matched PV and Pyr dendrites

```python
# Aggregate Pyr raw weights
raw_weight_path = "../data/weights_by_dendrite/depth_matched_pyr_raw"
files = os.listdir(raw_weight_path)

weights = []

for f in files:
    if 'npy' in f:
        w = np.load(os.path.join(raw_weight_path,f))
        weights.append(w[:,:7])

weights = np.concatenate(weights)

day = 1

pyr_rw = weights[:,day]
pyr_rw = pyr_rw[pyr_rw>0]
pyr_rw = np.log(pyr_rw)

#PV
del weights
raw_weight_path = "../data/weights_by_dendrite/depth_matched_pv_raw"
files = os.listdir(raw_weight_path)

weights=[]

for f in files:
    if 'npy' in f:
        w = np.load(os.path.join(raw_weight_path,f))
        weights.append(w[:,:7])

weights = np.concatenate(weights)

pv_rw = weights[:,day]
pv_rw = pv_rw[pv_rw>0]
pv_rw = np.log(pv_rw)

f,a = plt.subplots(1) 
f.set_size_inches(1.5*3,1.7*3)
a.plot(np.ones(pyr_rw.shape),pyr_rw,'o',markersize=20,alpha=0.1,fillstyle='none')
a.plot(np.ones(pv_rw.shape)*2,pv_rw,'ro',markersize=20,alpha=0.1,fillstyle='none')
a.plot([0.9,1.1],[np.median(pyr_rw),np.median(pyr_rw)],'c-')
a.plot([1.9,2.1],[np.median(pv_rw),np.median(pv_rw)],'m-')
a.set_xlim(0.5,2.5)

print('Pyr N Syns: ',pyr_rw.shape[0])
print('Pyr Median: ',np.median(pyr_rw))
print('Pyr SEM: ',scipy.stats.sem(pyr_rw))

print('')

print('PV N Syns: ',pv_rw.shape[0])
print('PV Median: ',np.median(pv_rw))
print('PV SEM: ',scipy.stats.sem(pv_rw))

print('')

print('Ranksum: {}'.format(scipy.stats.ranksums(pyr_rw,pv_rw)))
plt.show()
```

# Figure 2C-E

Comparison of distribution of weights between Pyr and PV dendrites

```python
day_to_plot = 5 # pick a single day

DATAPATH = "../data/weights_by_dendrite/norm_pyr/"
files = os.listdir(DATAPATH)

weights = []

for f in files:
    if 'npy' in f:
        w = np.load(os.path.join(DATAPATH,f))
        weights.append(w[:,:7])

weights = np.concatenate(weights, axis=0)
w = weights[:, day_to_plot]

w = np.squeeze(w[w > 0])

n_syn = w.shape[0]
binfactor = 7
n_bins = int(n_syn / binfactor)


f, a = plt.subplots(3, sharex=True)
f.set_size_inches(12,12)
a[0].hist(w, n_bins, density=True)

# Fitting
n_syn = w.shape[0]
print('N={}'.format(n_syn))

binfactor = 7
n_bins = int(n_syn / binfactor)

log_fit = fit_distribution(w, n_bins, "lognorm")
norm_fit = fit_distribution(w, n_bins, "norm")
logscale_fit = fit_distribution(np.log(w), n_bins, "norm")

params = logscale_fit[3]

arg = params[:-2]
loc = params[-2]
scale = params[-1]  # STD

a[0].plot(log_fit[1], "c", linewidth=3, label="lognorm")
a[0].plot(norm_fit[1], linewidth=3, color="grey")

a[1].plot(log_fit[2], log_fit[0], "c")
a[1].plot(norm_fit[2], norm_fit[0], color="grey")

a[2].plot(log_fit[2], np.power(log_fit[0], 2), "c")
a[2].plot(norm_fit[2], np.power(norm_fit[0], 2), color="grey")
plt.show()
```

```python
f, a = plt.subplots(1)
f.set_size_inches(12,3.9)
hist_info = a.hist(np.log(w), n_bins, color="blue", density=True)
a.plot(logscale_fit[1], "c")
a.set_xlim(-3,3)
a.set_yticks([0,0.25,0.5])
plt.show()
```

```python
f, a = plt.subplots(1)
f.set_size_inches(10,10)
sm.qqplot(
    np.log(w), scipy.stats.norm, distargs=arg, loc=loc, scale=scale, line="45", ax=a
)

a.set_aspect(1)

plt.show()
```
