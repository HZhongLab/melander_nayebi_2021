# Figure 1D: Pyr Correlations

```python
pyr_path = os.path.join(DATAPATH, "uncaging/punctaDfPyrRC.mat")
pyr_edata = load_ephys_data(pyr_path)

tifpath = os.path.join(DATAPATH, "uncaging/raw/pyr/pyr_005.tif")
synpath = os.path.join(DATAPATH, "uncaging/raw/pyr/Pyr_005.syn.mat")

dendrite = pyr_edata[4]

amps = []
idxs = []
on_dend = []

for count,idx in enumerate(dendrite.ephysTag):  # Only use indices meant for analysis
    idx = idx - 1

    if dendrite.punctaTag[count]==0: 
        on_dend.append(idx)

    idxs.append(idx)
    amp = dendrite.spotData[idx].ephysParameters2.amplitudeSlide[0]
    amps = np.append(amps, amp)

syn_data = load_synfile(synpath)
weights = extract(
    tifpath, synpath, bleedthrough=2, search_depth=5, integrate_over=5
)

weights = weights[idxs]

weights = weights / np.mean(weights)

fig, ax = plt.subplots(1)
fig.set_size_inches(9,9)
ax.plot(weights, amps, "go",markersize=20,fillstyle='none')
ax.set_title('N: {}, r: {}'.format(len(weights),scipy.stats.pearsonr(weights, amps)[0]))
ax.set_ylim(0)

x_vals = np.array([0, np.max(weights)])
fitline = scipy.stats.linregress(weights, amps)
slope = fitline[0]
intercept = fitline[1]
y_vals = x_vals * slope + intercept
ax.plot(x_vals, y_vals, "k--")

ax.plot(weights[on_dend],amps[on_dend],'ro',markersize=20,fillstyle='none')

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.set_ylim(0,np.max(amps)+3)
ax.set_xlim(0,np.max(weights)+0.2)
ax.set_xlabel('Normalized Weights')
ax.set_ylabel('Amplitudes')


plt.show()
```

# Figure 1F: PV and Pyr Correlation Across Dendrites

```python
# Aggreagate r-values across Pyr dendrites
pyr_rs = []
for count, ID in enumerate(pyr_IDs):
    tifpath = os.path.join(DATAPATH, "uncaging/raw/pyr/pyr_{}.tif".format(ID))
    synpath = os.path.join(DATAPATH, "uncaging/raw/pyr/Pyr_{}.syn.mat".format(ID))

    if ID == '009': # Experimenter noted that this cell seemed unhealthy - omit from experiments
        dendrite = pyr_edata[count+1] 
    else:
        dendrite = pyr_edata[count]

    amps = []
    idxs = []
    on_dend = []

    for count,idx in enumerate(dendrite.ephysTag):  # Only use indices meant for analysis
        idx = idx - 1

        idxs.append(idx)
        amp = dendrite.spotData[idx].ephysParameters2.amplitudeSlide[0]
        amps = np.append(amps, amp)

    syn_data = load_synfile(synpath)
    weights = extract(
        tifpath, synpath, bleedthrough=2, search_depth=5, integrate_over=5
    )

    weights = weights[idxs]
    weights = weights / np.mean(weights)

    r = scipy.stats.pearsonr(weights,amps)[0]
    pyr_rs.append(r)


# Aggregate r-values across PV dendrites
pv_rs = []
for count, ID in enumerate(pv_IDs):
    tifpath = os.path.join(DATAPATH, "uncaging/raw/pv/PV_{}.tif".format(ID))
    synpath = os.path.join(DATAPATH, "uncaging/raw/pv/PV_{}.syn.mat".format(ID))

    dendrite = pv_edata[count]

    amps = []
    idxs = []
    on_dend = []

    for count,idx in enumerate(dendrite.ephysTag):  # Only use indices meant for analysis
        idx = idx - 1
        idxs.append(idx)

        amp = dendrite.spotData[idx].ephysParameters2.amplitudeSlide
        if str(amp).lower()=='nan':
            amp = 0.0
        elif len(amp)==2:
            amp = amp[0]
        else:
            pass
        amps = np.append(amps, amp)

    syn_data = load_synfile(synpath)
    weights = extract(
        tifpath, synpath, bleedthrough=2, search_depth=5, integrate_over=5
    )

    weights = weights[idxs]
    weights = weights / np.mean(weights)

    r = scipy.stats.pearsonr(weights,amps)[0]
    pv_rs.append(r)
```

```python
pv_mean= np.mean(pv_rs)
pv_sem = scipy.stats.sem(pv_rs)

pyr_mean = np.mean(pyr_rs)
pyr_sem = scipy.stats.sem(pyr_rs)

print('Pyr Mean and SEM: {} +- {}'.format(pyr_mean,pyr_sem))
print('PV Mean and SEM: {} +- {}'.format(pv_mean,pv_sem))



fig,ax = plt.subplots(1)
fig.set_size_inches(1.5*3,1.7*3)

ax.plot(np.ones(len(pyr_rs)),pyr_rs,'bo',fillstyle='none',ms=23,lw=0.9)
ax.plot(np.ones(len(pv_rs))*2,pv_rs,'ro',fillstyle='none',ms=23,lw=0.9)

ax.plot([0.9,1.1],[pyr_mean,pyr_mean],'c',linewidth=2)
ax.plot([1.9,2.1],[pv_mean,pv_mean],'m',linewidth=2)

ax.set_ylim(0,1.01)
ax.set_xlim(0.5,2.5)

ax.set_yticks([0,0.5,1.0])

plt.show()
```

# Figure 1G: Spine vs PSD-95 Comparison

```python
sys.path.append('../data/misc')
from spine_roi_indexes import spine_indexes

pyr_path = os.path.join(DATAPATH, "uncaging/punctaDfPyrRC.mat")
pyr_edata = load_ephys_data(pyr_path)

green_rs = []
red_rs = []

n_syn = 0
for count, ID in enumerate(pyr_IDs):
    tifpath = os.path.join(DATAPATH, "uncaging/raw/pyr/pyr_{}.tif".format(ID))

    # Different syn files, drew ROIs in _red_ that more accurately capture spine
    red_synpath = os.path.join(DATAPATH, "uncaging/raw/pyr/Pyr_red_{}.syn.mat".format(ID)) 
    green_synpath = os.path.join(DATAPATH, "uncaging/raw/pyr/Pyr_{}.syn.mat".format(ID))


    if ID == '009':
        dendrite = pyr_edata[count+1] # Bart omitted second to last dendrite in his script, I do so here
    else:
        dendrite = pyr_edata[count]

    spine_idxs = copy.deepcopy(spine_indexes[ID])
    spine_idxs = np.array(spine_idxs)-1

    amps = []
    idxs = []
    on_dend = []

    # print(dendrite.punctaTag)
    for count,idx in enumerate(dendrite.ephysTag):  # Only use indices meant for analysis
        idx = idx - 1

        if idx in spine_idxs:
            idxs.append(idx)
            amp = dendrite.spotData[idx].ephysParameters2.amplitudeSlide[0]
            amps = np.append(amps, amp)

            n_syn += 1

    green_weights = extract(
        tifpath, green_synpath, bleedthrough=2, search_depth=5, integrate_over=5,color='green'
    )
    green_weights = green_weights[idxs]

    red_weights = extract(
        tifpath, red_synpath, bleedthrough=2, search_depth=5, integrate_over=5,color='red'
    )
    red_weights = red_weights[idxs]


    green_r = scipy.stats.pearsonr(green_weights,amps)[0]
    green_rs.append(green_r)

    red_r = scipy.stats.pearsonr(red_weights,amps)[0]
    red_rs.append(red_r)
    
red_mean = np.mean(red_rs)
green_mean = np.mean(green_rs)
red_sem = scipy.stats.sem(red_rs)
green_sem = scipy.stats.sem(green_rs)

print('Green: {} {}'.format(green_mean,green_sem))
print('Red: {} {}'.format(red_mean,red_sem))
print('Num Syn: {}'.format(n_syn))
w,p = scipy.stats.wilcoxon(x=green_rs,y=red_rs,alternative='greater')
print('Wilcoxon Statistic: ',w,' / P Value: ',p)

fig,ax = plt.subplots(1)
fig.set_size_inches(1.5*3,1.7*3)

for z in zip(green_rs,red_rs):
    ax.plot(z,'ko-',linewidth=1,fillstyle='full',ms=20,markerfacecolor='white')
ax.plot([-.1,0.1],[np.mean(green_rs), np.mean(green_rs)],'g-')
ax.plot([0.9,1.1],[np.mean(red_rs), np.mean(red_rs)],'r-')

ax.set_xticks([0,1])
ax.set_xticklabels(['PSD95','Spine V.'])
ax.set_ylim(0,1.0)
ax.set_xlim(-0.5,1.5)
ax.set_yticks([0,0.5,1.0])

plt.show()
```
